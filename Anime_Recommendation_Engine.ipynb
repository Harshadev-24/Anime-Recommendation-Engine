```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv("/content/anime.csv")
```

```python
features = ['Name','Genres','Type','Episodes', 'Studios', 'Source', 'Rating', 'Ranked', 'Popularity']
```

```python
def combine_features(row):
    return str(row['Name'])+" "+str(row['Genres'])+" "+str(row['Type'])+" "+str(row['Episodes'])+" "+str(row['Studios'])+" "+str(row['Source'])+" "+str(row['Rating'])+" "+str(row['Ranked'])+" "+str(row['Popularity'])
```

```python
for feature in features:
    df[feature] = df[feature].fillna('') #filling all NaNs with blank string

df["combined_features"] = df.apply(combine_features,axis=1) #applying combined_features() method over each rows of dataframe and storing the combined string in "combined_features" column
```

```python
df.iloc[0].combined_features
```

```python
cv = CountVectorizer() #creating new CountVectorizer() object
count_matrix = cv.fit_transform(df["combined_features"]) #feeding combined strings(anime contents) to CountVectorizer() object
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["Name"].values[0]
def get_index_from_title(Name):
    return df[df.Name == Name].index[0]
```

```python
anime_user_likes = "Air"
anime_index = get_index_from_title(anime_user_likes)
similar_animes = list(enumerate(cosine_sim[anime_index])) #accessing the row corresponding to given movie to find all the similarity scores for that movie and then enumerating over it
```

```python
sorted_similar_animes = sorted(similar_animes,key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
print("Top 5 similar animes to "+anime_user_likes+" are:\n")
for element in sorted_similar_animes:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>5:
        break
```
